```python
def remove_punctuation(txt):
    txt_nopunct = "".join([c for c in txt if c not in string.punctuation])
    return txt_nopunct
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(analyzer = 'word',ngram_range = (1,3),min_df = 0,stop_words = 'english')
```

```python
fd_tidf = tf.fit_transform(d_table['cleaned'])
fd_tidf.shape
```

```python
tf.get_feature_names()
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(fd_tidf,fd_tidf)
#cosine_sim
```

```python
fd_index = pd.Series(d_table.index,index = d_table['title'])
fd_index
```

```python
def ingredient_similarity(title) :
    index = fd_index[title]
    ingd_list = list(enumerate(cosine_sim[index]))
    ingd_list = sorted(ingd_list,key = lambda x : x[1],reverse =True)
    ingd_list = ingd_list[1:10]
    ingd_list = [i[0] for i in ingd_list]
    return d_table['title'].iloc[ingd_list]
```
