```python
food_dset['ingredients'] = food_dset.ingredients.str.replace('$,(){}%', '')
print(food_dset['ingredients'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
tf = TfidfVectorizer(analyzer = 'word',ngram_range = (1,2),min_df = 0,stop_words = 'english')
```

```python
food_tfidf = tf.fit_transform(food_dset['ingredients'])
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(food_tfidf,food_tfidf)
print(food_tfidf)
```

```python
#food_dset['fdc_id'] = food_dset.fdc_id.astype(object)
food_index = pd.Series(food_dset.index,index = food_dset['brand_owner'])
#movie_index = pd.Series(movie_dset.index,index = movie_dset['title'])
print(food_dset['brand_owner'])
```

```python
def food_recommendation(brand_owner) :
    index = food_index[brand_owner]
    food_list = list(enumerate(cosine_sim[index]))
    food_list = sorted(food_list,key = lambda x : x[1],reverse =True)
    food_list = food_list[1:5]
    food_list = [i[0] for i in food_list]
    return food_dset['brand_owner'].iloc[food_list]
```
