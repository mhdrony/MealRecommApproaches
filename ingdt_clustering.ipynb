```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("C:\\Users\\non_coder\\Desktop\\datasets_fldr\\exported_dta.csv", nrows=1000)
data.head(10)
```

```python
data["cleaned"].fillna('',inplace = True)
```

```python
data = pd.DataFrame(data)
```

```python
pd.set_option('display.max_colwidth', 100)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(analyzer = 'word',ngram_range = (1,4),min_df = 0,stop_words = 'english')
```

```python
fd_tfidf = tf.fit_transform(data['cleaned'])
```

```python
tf.get_feature_names()
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(fd_tfidf,fd_tfidf)
#cosine_sim
```

```python
fd_index = pd.Series(data.index,index = data['title'])
fd_index
```

```python
def ingredient_similarity(title) :
    index = fd_index[title]
    ingd_list = list(enumerate(cosine_sim[index]))
    ingd_list = sorted(ingd_list,key = lambda x : x[1],reverse =True)
    ingd_list = ingd_list[1:10]
    ingd_list = [i[0] for i in ingd_list]
    return data['title'].iloc[ingd_list]
```
