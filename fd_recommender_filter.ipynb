```
tmp = "fats  sodium  added sugars  vegetables  fruits  whole grain  lean meat  fat free dairy  fat free milk  beans  nuts animal fat  butter salt peanut   lemon  milk unsweetened  bananas raw  meatless cheese  added vitamin   almonds  sugars granulated  vitamin  dry wheat  dry water  almonds seeds  wheat flour  spices garlic  raw rice  dates  honey  oats  chicken  calcium sulfate carrots  cashew nuts  broccoli  cooking water  tomatoes  coriander seed  rice white bread  eggs  saturated trans fat"
similarity = []
for i in sim_fdlist.cleaned:
        #print(i)
        #print('\n')
        ratio = process.extract( i, suggested_ingds, limit=1)
        #print(suggested_ingds)
        #print('\n')
        #actual_fdlist.append(ratio[0][0])
        Ratio = fuzz.ratio(i,tmp)
        #print(Ratio)
        #print('\n')
        similarity.append(Ratio)
        #print(similarity)
#d_table['actual_fdlist'] = pd.Series(actual_fdlist)
#print(pd.Series(similarity))
sim_fdlist['similarity'] = pd.Series(similarity)
sim_fdlist.rename(columns={'cleaned':'ingredients', 'similarity':'SM_ingd'},inplace=True)
#ingred_list = sim_fdlist['similarity']
#ingred_list
sim_fdlist
```
